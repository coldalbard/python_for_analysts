```python
import pandas as pd
import matplotlib.pyplot as plt
house = pd.read_csv('/gb/analitcPython/homework/homework4/kc_house_data.csv', sep=',', encoding='windows-1251')
laptops = pd.read_csv('/gb/analitcPython/homework/homework4/laptop_price.csv', sep=',', encoding='windows-1251')
```

### 2.1. Изучите распределение домов от наличия вида на набережную

```python
waterfront = house['waterfront'].value_counts()
plt.pie(waterfront, autopct='%1.1f%%', labels=["Нет", "Да"])
plt.title('Распределение домов от наличия вида на набережную')
```

### 2.2. Изучите распределение этажей домов

```python
plt.hist(house['floors'])
plt.xlabel('Этажи')
plt.ylabel('Количество')
plt.title('Гистограмма этажей');
```

### 2.3. Изучите распределение состояния домов

```python
plt.hist(house['condition'])
plt.xlabel('Состояние домов')
plt.ylabel('Количество')
plt.title('Гистограмма состояния домов');
```
